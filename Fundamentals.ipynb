```python
import yfinance as yf
import pandas as pd
import os
ticker_input = "AUR, AVTR, AZTA, CELH, DHX, DIOD, EVTL, GSAT, INTC, MXL, SATS, SEDG, SERV, SOFI, SVCO, TSLA, VEON, WRD"

tickers = [t.strip().upper() for t in ticker_input.split(',')]

data = []
for ticker in tickers:
    stock = yf.Ticker(ticker)
    info = stock.info
    data.append({
        'Ticker': ticker,
        'Trailing P/E': info.get('trailingPE', 'N/A'),
        'Forward P/E': info.get('forwardPE', 'N/A'),
        'Beta': info.get('beta', 'N/A'),
        '好看吗': '',
        '等什么': ''
    })

df = pd.DataFrame(data)

df.to_csv(os.path.expanduser('~/Desktop/Portfolio_Quality.csv'))


df

```
